```python
import numpy as np
import pandas as pd
import anndata as ad
import os
from joblib import load, dump
from scipy import sparse

pth='/mnt/d/temp/ywt01_1357/processed'
identifier='bardensr_5px_001_005'
df=load(os.path.join(pth,identifier,'filt_neurons.joblib'))
d=df['filt_neurons'].copy()
# --- X matrix ---
X = d["expmat"]

if sparse.issparse(X):
    X = X.tocsr()

# --- obs dataframe ---
obs = pd.DataFrame({
    "cell_id": np.asarray(d["id"]).ravel(),
    "pos_x": np.asarray(d["pos10x_x"]).ravel(),
    "pos_y": np.asarray(d["pos10x_y"]).ravel(),
    "slice": np.asarray(d["slice"]).ravel(),
})

# --- var dataframe ---
genes=[g[0][0].item() for g in d['genes']]
genes = np.array(genes)

var = pd.DataFrame(index=genes)
var["gene_symbol"] = genes

# --- make AnnData ---
adata = ad.AnnData(X=X, obs=obs, var=var)

# optional: put spatial coordinates in obsm like Scanpy expects
adata.obsm["spatial"] = obs[["pos_x", "pos_y"]].to_numpy()

print(adata)
print(adata.obs.head())
print(adata.var.head())

adata.write_h5ad(os.path.join(pth,f'filt_neuron_{identifier}.h5ad'))
```
